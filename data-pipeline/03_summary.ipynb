# Group by

```python
df.groupby("portno")
```

```python
"""
SELECT
    portno
    , COUNT(1)
FROM table
GROUP BY portno, eff_dt
"""
df.groupby(["portno", "eff_dt"]).size()
```

```python
"""
select
    portno
    , eff_dt
    , sum(percent_tna) as percent_tna
    , sum(nbr_shares) as nbr_shares
from df
group by portno, eff_dt
"""
df.groupby(["portno", "eff_dt"]).sum()[
    ["percent_tna", "nbr_shares"]
]
```

```python
"""
select
    portno
    , eff_dt
    , sum(percent_tna) as percent_tna
    , sum(nbr_shares) as nbr_shares
from df
group by portno, eff_dt
having sum(percent_tna) > 0.5
"""
df_groupby = df.groupby(["portno", "eff_dt"]).sum()[
    ["percent_tna", "nbr_shares"]
]
filter_percent_tna_gt_05 = df_groupby["percent_tna"] > 0.5
df_groupby[filter_percent_tna_gt_05]
```

```python
"""
select
    portno
    , eff_dt
    , sum(percent_tna) as percent_tna
    , sum(nbr_shares) as nbr_shares
from df
group by portno, eff_dt
having sum(percent_tna) > 0.5
order by percent_tna desc
"""
df_groupby = df.groupby(["portno", "eff_dt"]).sum()[
    ["percent_tna", "nbr_shares"]
]
filter_percent_tna_gt_05 = df_groupby["percent_tna"] > 0.5
df_groupby[filter_percent_tna_gt_05].sort_values("percent_tna", ascending=False)
```

```python
"""
select
    portno
    , eff_dt
    , sum(percent_tna) as percent_tna
    , sum(nbr_shares) as nbr_shares
from df
group by portno, eff_dt
having sum(percent_tna) > 0.5
order by percent_tna desc, nbr_shares
"""
df_groupby = df.groupby(["portno", "eff_dt"]).sum()[
    ["percent_tna", "nbr_shares"]
]
filter_percent_tna_gt_05 = df_groupby["percent_tna"] > 0.5
df_groupby[filter_percent_tna_gt_05].sort_values(
    ["percent_tna", "nbr_shares"], ascending=[False, True],
)
```
