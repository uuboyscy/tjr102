```python
import pandas as pd

# Sample data: stock prices over 7 days
data = {
    'date': pd.date_range(start='2025-07-01', periods=7),
    'price': [100, 102, 101, 105, 107, 110, 108]
}

df = pd.DataFrame(data)

df
```

```python
# Calculate 3-day moving average
df['3_day_avg'] = df['price'].rolling(window=3, min_periods=2).mean()

df
```
